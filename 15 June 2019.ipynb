```python
def data(st):
    for i in range(0,len(st)):
        if(ord(st[i])>=65 and ord(st[i])<=90):
            print(chr(ord(st[i])+32),end='')
        elif(ord(st[i])>=97 and ord(st[i])<=122):
            print(chr(ord(st[i])-32),end='')
            
data("viJAy")
```
